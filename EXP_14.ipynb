```python
import pandas as pd
import numpy as np

data = {
    'Product' : ['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price' : [55000, 20000, 30000, 2000, 8000, 45000],
    'Units_sold' : [25, 65, 35, 80, 50, 15],
    'Discount' : [10, 15, 12, 20, 18, 8]
}

df = pd.DataFrame(data)
df
```

```python
#1. Min-Max Normalization
#Min-max normalization for price

df['Price_MinMax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())

print('\nMin-Max Normalization for Price')
print(df[['Product','Price', 'Price_MinMax']])
```

```python
# Normalize multiple columns
cols = ['Price', 'Units_sold', 'Discount']

df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())

print('\nNormalized dataset: ')
print(df_norm)
```

```python
#2. Z-score normalization

df['Units_sold_Zscore'] = (df['Units_sold'] - df['Units_sold'].mean()) / df['Units_sold'].std()

print('\nZ-score normalization for Units_sold')
print(df[['Product','Units_sold', 'Units_sold_Zscore']])
```

```python
#3. Decimal scaling
df['Price_Decimal'] = df['Price'] / 100000

print('\nDecimal scaling for Price')
print(df[['Product','Price', 'Price_Decimal']])
```

```python
# Turning categorical variables into quantitative variables

data = {
    'Order_ID' : [1001, 1002, 1003, 1004, 1005, 1006],
    'Customer_Gender' : ['Male', 'Female', 'Male', 'Female', 'Male', 'Male'],
    'Product_Category' : ['Electronics', 'Clothing', 'Home', 'Electronics', 'Beauty', 'Electronics'],
    'Payment_Method' : ['Credit Card', 'UPI', 'UPI', 'Credit Card', 'UPI', 'Credit Card'],
    'City' : ['Pune', 'Mumbai', 'Delhi', 'Bangalore', 'Pune', 'Hyderabad'],
    'Order_value' : [4500, 2300, 3200, 5200, 1800, 4100]
}

df = pd.DataFrame(data)
df
```

```python
# Method 1: Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender_Label'] = le.fit_transform(df['Customer_Gender'])
df
```

```python
# Method 2: One Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Product_Category'])
df_encoded
```
